```python
import pandas as pd
india_weather=[("Mumbai",32,80),
               ("Delhi",45,60),
               ("Banglore",30,78)]
df3=pd.DataFrame(india_weather,columns=["city","humidity","temperature"])
df3
```

```python
import pandas as pd
us_weather=[("New York",21,68),
               ("Chicago",14,65),
               ("Orland",35,75)]
df4=pd.DataFrame(us_weather,columns=["city","humidity","temperature"])
df4
```

# Row-wise Concatenation 

```python
import pandas as pd
frames = [df3, df4]

result = pd.concat(frames)
result
```

# If you want continuous index


```python
import pandas as pd
frames = [df3, df4]

result = pd.concat(frames,ignore_index=True)
result
```

# Column-Wise Concatenation

```python
import pandas as pd
frames = [df3, df4]

result = pd.concat(frames,axis=1)
result
```

```python
import pandas as pd
temperature=[("Mumbai",32),
               ("Delhi",45),
               ("Banglore",30),
               ("Hyderabad",42)]
df3=pd.DataFrame(temperature,columns=["city","temperature"])
df3
```

```python
import pandas as pd
humidity=[("Mumbai",68),
               ("Delhi",65),
               ("Banglore",75)]
df4=pd.DataFrame(humidity,columns=["city","humidity"])
df4
```

# Merging Dataframes

```python
import pandas as pd
#frames1 = [df3, df4]

result1 = pd.merge(df3,df4,on='city')
result1
```

# Outer Join

```python
import pandas as pd
#frames1 = [df3, df4]

result2 = pd.merge(df3,df4,on='city',how='outer')
result2
```
